# Representation Demystifies Flat Minima
## Notebook 37: Conclusions

This notebook synthesizes the results from the repository.

Rather than introducing new experiments, it asks:

> What survives representation?

The previous notebooks established:

- Hessians measure curvature.
- Curvature depends on coordinates.
- Equivalent computations can have different measured sharpness.
- Correlation is not automatically causation.

```python
import json
import zipfile
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

FIGURES = ROOT / "figures"
RESULTS = ROOT / "results"
DOWNLOADS = ROOT / "downloads"

for path in [FIGURES, RESULTS, DOWNLOADS]:
    path.mkdir(parents=True, exist_ok=True)
```

## 1. Classical hypothesis

A simplified version of the traditional flat-minima narrative:

> flatter minima generalize better.

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.axis("off")

ax.text(
    0.5,
    0.75,
    "Flat minima",
    ha="center",
    fontsize=22
)

ax.arrow(
    0.5,
    0.60,
    0,
    -0.20,
    width=0.005,
    length_includes_head=True
)

ax.text(
    0.5,
    0.25,
    "Better generalization",
    ha="center",
    fontsize=22
)

ax.set_title("Classical hypothesis")

fig_path = FIGURES / "37_classical_hypothesis.png"
plt.savefig(fig_path, bbox_inches="tight", dpi=180)
plt.show()

print("Saved:", fig_path)
```

## 3. Same computation, different geometry

This is the central observation of the repository.

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.axis("off")

ax.text(
    0.25,
    0.75,
    "Representation A\n\nH = 0.04\n\nPrediction = y=x",
    ha="center",
    va="center",
    fontsize=16,
    bbox=dict()
)

ax.text(
    0.75,
    0.75,
    "Representation B\n\nH = 10.72\n\nPrediction = y=x",
    ha="center",
    va="center",
    fontsize=16,
    bbox=dict()
)

ax.text(
    0.5,
    0.20,
    "Same computation\nDifferent geometry",
    ha="center",
    fontsize=20
)

fig_path = FIGURES / "37_same_computation_different_geometry.png"
plt.savefig(fig_path, bbox_inches="tight", dpi=180)
plt.show()

print("Saved:", fig_path)
```

## 4. What survives?

Distinguish computation-level properties from representation-level properties.

```python
survives = pd.DataFrame([
    ["✓", "Function"],
    ["✓", "Predictions"],
    ["✓", "Test error"],
    ["✗", "Sharpness"],
    ["✗", "Coordinates"],
    ["✗", "Parameter norm"],
], columns=["Survives?", "Property"])

survives
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.axis("off")

table = ax.table(
    cellText=survives.values,
    colLabels=survives.columns,
    loc="center",
    cellLoc="left",
    colLoc="left"
)

table.auto_set_font_size(False)
table.set_fontsize(14)
table.scale(1, 1.8)

ax.set_title("Properties that survive representation changes")

fig_path = FIGURES / "37_surviving_properties.png"
plt.savefig(fig_path, bbox_inches="tight", dpi=180)
plt.show()

print("Saved:", fig_path)
```

## 5. Representation versus computation

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.axis("off")

ax.text(
    0.25,
    0.75,
    "Representation",
    ha="center",
    fontsize=18
)

ax.text(
    0.25,
    0.50,
    "Geometry",
    ha="center",
    fontsize=18
)

ax.text(
    0.25,
    0.25,
    "Measurements",
    ha="center",
    fontsize=18
)

ax.text(
    0.75,
    0.75,
    "Computation",
    ha="center",
    fontsize=18
)

ax.text(
    0.75,
    0.50,
    "Predictions",
    ha="center",
    fontsize=18
)

ax.text(
    0.75,
    0.25,
    "Performance",
    ha="center",
    fontsize=18
)

ax.arrow(0.25,0.68,0,-0.10,width=0.004)
ax.arrow(0.25,0.43,0,-0.10,width=0.004)

ax.arrow(0.75,0.68,0,-0.10,width=0.004)
ax.arrow(0.75,0.43,0,-0.10,width=0.004)

ax.set_title("Representation versus computation")

fig_path = FIGURES / "37_representation_vs_computation.png"
plt.savefig(fig_path, bbox_inches="tight", dpi=180)
plt.show()

print("Saved:", fig_path)
```

## 6. Future question

Instead of asking:

> Can flatness explain generalization?

ask:

> Can we build representation-invariant measures of generalization?

This shifts attention from coordinate-dependent measurements
toward invariant properties of computation.

## Engineering statement

Curvature measures geometry.

Geometry depends on representation.

Computation survives representation.

Scientific explanations should identify what survives.

```python
summary = {
    "title": "Conclusions",
    "core_claim": "Computation survives representation.",
    "future_question": "Can we build representation-invariant measures of generalization?",
    "engineering_statement":
        "Curvature measures geometry. Geometry depends on representation. Computation survives representation. Scientific explanations should identify what survives."
}

summary_json = RESULTS / "37_conclusions_summary.json"

with open(summary_json, "w") as f:
    json.dump(summary, f, indent=2)

summary_df = pd.DataFrame([summary])
summary_csv = RESULTS / "37_conclusions_summary.csv"
summary_df.to_csv(summary_csv, index=False)

print(summary_json)
print(summary_csv)
```
