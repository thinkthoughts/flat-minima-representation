```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

FIGURES = Path('../figures')
FIGURES.mkdir(exist_ok=True)
```

## A simple minimum

$L(x)=x^2$

```python
x=np.linspace(-3,3,500)
loss=x**2
plt.figure(figsize=(8,4))
plt.plot(x,loss)
plt.grid(True)
plt.savefig(FIGURES/'00_simple_minimum.png',bbox_inches='tight')
plt.show()
```

## Flat versus sharp

```python
x=np.linspace(-1,1,500)
flat=x**2
sharp=100*x**2
plt.figure(figsize=(8,4))
plt.plot(x,flat,label='Flat')
plt.plot(x,sharp,label='Sharp')
plt.ylim(0,25)
plt.legend()
plt.grid(True)
plt.savefig(FIGURES/'00_flat_vs_sharp.png',bbox_inches='tight')
plt.show()
```
